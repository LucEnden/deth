```python
import json
import os

raw_data_path = os.path.abspath('./raw_platinum_trainers.json')
parsed_data_path = os.path.abspath('./parsed_platinum_trainers.json')
```

https://github.com/OttoTonsorialist/pkmn_gen_four_trainer_data

```python
import requests

data_url = 'https://raw.githubusercontent.com/OttoTonsorialist/pkmn_gen_four_trainer_data/refs/heads/main/platinum_trainers.json'
response = requests.get(data_url)

if not os.path.exists(raw_data_path):
    with open(raw_data_path, 'wb') as file:
        file.write(response.content)
```

```python
if not os.path.exists(parsed_data_path):
    with open(parsed_data_path, 'w') as parsed_file:
        with open(raw_data_path, 'r') as file:
            parsed_file.write(json.load(file)['trainers'])
```

## Missing rival battle locations

```python
with open(parsed_data_path, 'r') as file:
    trainers = json.load(file)

idx_of_trainers_missing_location = [
    trainer['rom_id']
    for trainer in trainers[1:]
    if not trainer['trainer_location'] and trainer['trainer_name'] == 'Rival'
]

for idx in idx_of_trainers_missing_location:
    querry = 'https://www.google.com/search?q='
    querry += trainers[idx]['trainer_name'].replace(' ', '+') + '+'
    querry += '+'.join([
        f"Lv.+{p['level']}+{p['species']}" for p in trainers[idx]['pokemon']
    ])
    querry += '+location+platinum'
    print(idx, querry.strip())
```
